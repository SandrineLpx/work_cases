```python
# In this script we cover k-means clustering in Python

#############################
# Load the iris dataset
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.cluster import KMeans

# Load iris data
iris = load_iris()
data = iris.data  # first 4 columns (features)
target = iris.target  # species labels

# Create a pandas DataFrame for better inspection
df = pd.DataFrame(data, columns=iris.feature_names)
df['Species'] = target

# Look at the structure
print(df.head())
print(df.info())

########################
## K-MEANS CLUSTERING ##
########################

# Initialize K-Means with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)

# Fit the model on feature columns (exclude Species)
kmeans.fit(data)

# Get cluster assignments
IrisKMCGroups = kmeans.labels_

# Compare predicted clusters with actual species
print("\nCluster vs Actual Species:")
print(pd.crosstab(df['Species'], IrisKMCGroups))

# Show centers of each cluster
print("\nCluster Centers:")
print(kmeans.cluster_centers_)
```
